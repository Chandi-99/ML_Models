```python
#import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
#import csv file
wineData = pd.read_csv("C:\\Users\\Chandi\\Desktop\\Data\\wine.csv")
wineData
```

```python
#Give names to columns according to the data set
colnames=['Class', 'Alcohol', 'Malic acid', 'Ash','Alcalinity of ash','Magnesium','Total phenols','Flavanoids','Nonflavanoid phenols','Proanthocyanine','Color Intensity','Hue','OD280/OD315 of diluted wines','Proline'] 
wineData = pd.read_csv('C:\\Users\\Chandi\\Desktop\\Data\\wine.csv', names=colnames, header=None)
wineData
```

```python
#Identify ranges of classes in the data set
wineData[wineData.Class==2].head()
```

```python
wineData[wineData.Class==3].head()
```

```python
#Classify class 1 = wine 1, class 2 = wine 2, class 3 = wine 3:
wines = ["wine 1", "wine 2", "wine 3"]
wineData['Wine_name'] = wineData.Class.apply(lambda x:wines[x-1] )
wineData['Wine_name']
```

```python
#Interpret classes
wine1Data = wineData[:59]
wine2Data = wineData[59:130]
wine3Data = wineData[130:]
```

```python
#dimensionality reduction
X = wineData.drop(['Class','Wine_name'], axis='columns')
Y = wineData.Class
for testSize in range(1,6):
    print("test size =", testSize/10)
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=(testSize/10), random_state = 10)
    len(X_train)
    len(X_test)
    #classify data set using KNN
    accuracy = 0
    for x in range(1,6):
        knn = KNeighborsClassifier(n_neighbors=x)
        knn.fit(X_train, Y_train)
        temp = knn.score(X_test, Y_test)
        if temp > accuracy:
            N = x
            accuracy = temp
    print("Number of Neighbours :", N)
    print("Accuracy :", accuracy,"\n")
```
